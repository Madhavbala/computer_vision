```python
# Example 2
faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('yogiji.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

face = faceCascade.detectMultiScale(img_gray, 1.2,4)
(x,y,w,h) = face[0]

cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 6)

plt.figure(figsize = (10,10))
plt.imshow(img)
plt.axis('off')
```

```python
# Example 3 (multiple faces)
faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('team.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

faces = faceCascade.detectMultiScale(img_gray, 1.2, 4)
faces
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 6)
    
plt.figure(figsize = (14,10))
plt.imshow(img)
plt.axis('off')
```

```python
# Function to to detect face
def detect_face(image_path):
    '''
    Input : Path to an image
    Return: Image with bounding boxes around faces
    '''
    
    img = cv2.imread(image_path)                                         # Read the image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                           # Convert to RGB
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)                     # Create a grayscale copy
    
    # Create a cascade classifier and use detectMultiScale function
    faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = faceCascade.detectMultiScale(img_gray, 1.2, 4)
    
    for x,y,w,h in faces:
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 6)
        
    return img
```

```python
image = detect_face('modiji.jpg')

plt.figure(figsize = (10,10))
plt.imshow(image)
plt.axis('off')
```

```python
image = detect_face('team.jpg')

plt.figure(figsize = (10,10))
plt.imshow(image)
plt.axis('off')
```
