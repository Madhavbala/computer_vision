```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# Reading an image
panda = cv2.imread('panda.jpg')
plt.imshow(panda)
plt.axis('off')
```

```python
panda.shape
```

```python
koala = cv2.imread('koala.jpg')
plt.imshow(koala)
plt.axis('off')
```

```python
# Deafault format for cv2 in BGR
# Changing color format from BGR to RGB
koala_orig = cv2.cvtColor(koala, cv2.COLOR_BGR2RGB)
plt.imshow(koala_orig)
plt.axis('off')
```

```python
# Converting an image to grayscale
panda_gray = cv2.cvtColor(panda, cv2.COLOR_BGR2GRAY)
plt.imshow(panda_gray, cmap = 'gray')
plt.axis('off')
```

```python
koala_gray = cv2.cvtColor(koala, cv2.COLOR_BGR2GRAY)
plt.imshow(koala_gray, cmap='gray')
plt.axis('off')
```

```python
# Extracting edges / pattern
panda2 = cv2.Canny(panda, 150,200)
plt.imshow(panda2, 'gray')
plt.axis('off')
```

```python
koala2 = cv2.Canny(koala, 150,200)
plt.imshow(koala2, 'gray')
plt.axis('off')
```

```python
# Highlight the edges
panda3 = cv2.dilate(panda2, kernel = (3,3), iterations = 5)
plt.imshow(panda3, 'gray')
plt.axis('off')
```

## Resizing and Cropping

```python
panda = cv2.imread('panda.jpg')
plt.imshow(panda)
plt.axis('off')
```

```python
panda.shape
```

```python
panda2 = cv2.resize(panda, (2000,1300))
plt.imshow(panda2)
plt.axis('off')
```

```python
panda2.shape
```

```python
panda2 = cv2.resize(panda, (2000,2000))
plt.imshow(panda2)
plt.axis('off')
```

```python
panda2.shape
```

```python
panda2 = cv2.resize(panda, (100,70))
plt.imshow(panda2)
plt.axis('off')
```

```python
panda2 = cv2.resize(panda, (3000, 2000))
cv2.imwrite('panda_resized.png', panda2)
```

```python
# Cropping
plt.imshow(panda, 'gray')
plt.axis('off')
```

```python
panda_cropped = panda[100:800, 300:1100]
plt.imshow(panda_cropped)
plt.axis('off')
```

## Drawing a bounding box (rectangle)

```python
panda = cv2.imread('panda.jpg')
plt.imshow(panda)
plt.axis('off')
```

```python
cv2.rectangle(panda, (200,100),(1100,700), (255,0,0), 10)
plt.imshow(panda)
plt.axis('off')
```

```python
plt.figure(figsize = (14,14))
plt.imshow(panda)
plt.axis('off')
```

## Writing text on images

```python
plt.figure(figsize  = (14,14))
koala = cv2.imread('koala.jpg')
koala = cv2.cvtColor(koala, cv2.COLOR_BGR2RGB)
cv2.putText(koala, 'What are you looking at, let me have my food', (100,600), cv2.FONT_HERSHEY_COMPLEX, 1.2, (255,0,0), 3)
plt.imshow(koala)
plt.axis('off')
cv2.imwrite('koala_meme.png', koala)
```
